```python
import numpy as np
import pandas as pd
pd.options.display.max_columns = 200

gamelogsDF = pd.read_csv("GL2016.TXT",header=None)
# fix column names
gamelogsDF.columns = range(1,162)
```

```python
# COLUMN-WISE and row-wise operations

# columns 10 and 11 are the number of runs for visiting and home teams respectively
gamelogsDF[[10,11]].head()
```

```python
# how many runs were scored by the visiting and home teams in all the games?
gamelogsDF[[10,11]].sum(axis=0)
```

```python
# we added up all the values in column 10 and column 11 for all the rows and we see that
# the home teams outscored the visiting teams
gamelogsDF[[10,11]].sum(axis=1)
```

```python
# let's make a new column to store this information
gamelogsDF["totalRuns"] = gamelogsDF[[10,11]].sum(axis=1)
gamelogsDF.head()
```

```python
# what were the games with the highest number of runs scored?
gamelogsDF.sort_values("totalRuns",ascending=False).head(10)
```

```python
%matplotlib inline
# let's plot a distribution of the total runs scored
gamelogsDF["totalRuns"].hist(bins=30)
```

```python
# it looks like 9 runs happens the most frequently.  let's verify this with value_counts
gamelogsDF["totalRuns"].value_counts()
```

```python
# let's look at seaborn's histogram
import seaborn as sb
sb.distplot(gamelogsDF[[10,11]].sum(axis=1),bins=30,kde=False)
# same as sb.distplot(gamelogsDF["totalRuns"],bins=30)
# remember: histograms are special kinds of bar charts that allow us to see the frequency of
# occurrences 
```

```python
# apply and lambda functions
# how do we run functions on individual values in a dataframe?

# column 81 is the 1B umpire's name for each game
gamelogsDF[81]
```

```python
# let's say i have a (ridiculous) hypothesis about the length (number of characters)
# of the 1B umpire's name how would we get the length of all of the umpire's names?
```

```python
len(gamelogsDF[81]) # this gives you the number of rows
```

```python
gamelogsDF[81].apply(len)
# the apply method applies the function passed as a parameter to it to each item in column 81
```

```python
# what is the average number of characters in 1st base umpires' names?
np.mean(gamelogsDF[81].apply(len))
```

```python
gamelogsDF[81].apply(len).hist(bins=11)
# distribution of 1B umpire name length
```

```python
# what if i wanted to find the average number of vowels in the first base umpires' names?
# to do this, we'd need to count up the number of vowels in each umpire's name and then take
# the average
import re
umpireName = "Joe Smith"
umpireVowels = re.findall("[aeiou]",umpireName.lower())
print(len(umpireVowels))
```

```python
# we'd need to do this for all umpires. we can either write a function ourselves to do it
# or we can use what's called a lambda function with apply.

# a lambda function is an anonymous function -- it is not named and used one time, typically
# for a small task.  here's how it works:
vowelCount = lambda x: len(re.findall("[aeiou]",x.lower()))
print(vowelCount(umpireName))
```

```python
# let's use a lambda function with apply on column 81
gamelogsDF[81].apply(lambda x: len(re.findall("[aeiou]",x.lower())))
```

```python
# what's the average number of vowels in a 1B umpire's name?
np.mean(gamelogsDF[81].apply(lambda x: len(re.findall("[aeiou]",x.lower()))))
```

```python
gamelogsDF[81].apply(lambda x: len(re.findall("[aeiou]",x.lower()))).hist(bins=5)
```
