### Part 2: Beautiful Soup

Using HouseSoup (from the class notes), print out ever house member's name, political party, and state abbreviation. 

For instance, Boulder's congressman should look like: Joe Neguse (D-CO)

```python
import requests
from bs4 import BeautifulSoup
houseXML = requests.get("http://clerk.house.gov/xml/lists/MemberData.xml").text
houseSoup = BeautifulSoup(houseXML,"lxml")
```

Some useful tags: 

official-name

state (you will have to get the postal-code from within this tag...)

party

```python
#how do we iterate through all members?
for member in houseSoup.members.contents:
    
    #how do we get a member's name?
    name = member.find("official-name").text
    
    #how do we get a member's party?
    party=member.find("party").text
    
    #how do we get a member's state abbreviation?
    abbr = member.find("state")['postal-code']
    
    #make sure to only print filled seats!
    if(name):
        
        #how do we format and print that info?
        print('%s (%s-%s)' %(name,party,abbr))
```
