## Sorting

Using the .sort_values() function, we can sort the dataframes by specified columns!

Options:
- df.sort_values(by=['col1'])
- df.sort_values(by=['col1', 'col2'])
- df.sort_values(by='col1', ascending=False)
- df.sort_values(by='col1', ascending=[booleans])

```python
# in this line I take the date, state, positive, negative columns of the covid df, sort it by the most postive values,
# and then use the index locator to get the 50th-100th indexes 

covidDF[["date","state","positive","negative"]].sort_values("positive",ascending=False).iloc[50:100]
```

```python
# we can see this is all ny for different days, so lets see if we cant get only unique states 
covidDF[["date","state","positive","negative"]].sort_values("positive",ascending=False).head(5)
```

```python
# Group the rows in the covid dataFrame by state
covidStateGrouping = covidDF[["date","state","positive","negative"]].sort_values("positive",ascending=False).groupby("state")
covidStateGrouping.groups

# lets create a new data frame of the max positive cases? or the sum?
maxCovidDatePerState = covidStateGrouping.agg({"positive":max}).sort_values("positive",ascending=False)
maxCovidDatePerState
```

```python
#Okay now lets look at spececifc states curve -- is it flattening??
covidStateGrouping.groups

# Now lets visualize this
import seaborn as sb
%matplotlib inline

covidStateGrouping.groups
covidStateGrouping.get_group('MA')[["date","positive"]]
```

```python
# lets play around 
#sb.distplot(covidStateGrouping.get_group('MA')[["date","positive"]])

# covidStateGrouping.get_group('MA')[["date","positive"]].hist

sb.factorplot(x="date",y="positive", data = covidStateGrouping.get_group('MA')[["date","positive"]], size = 12)
```

```python


#Goal: write a function that vizualises a states positive covid results for a distribution of dates 
def visualizeStatePositives(stateAbrev):
    print("Positive covid cases in "+stateAbrev+":")
    sb.factorplot(x="date",y="positive", data = covidStateGrouping.get_group(stateAbrev)[["date","positive"]], size = 8)
```

```python
visualizeStatePositives('AL')
```
