```python
import numpy as np
import pandas as pd 
pd.options.display.max_columns = 200 

gamelogsdf = pd.read_csv("GL2016.TXT",header=None)
```

```python
gamelogsdf.columns = range(1,162)
gamelogsdf.head(10)
```

```python
#Column wise and row wise operations

#column 10 and 11 are runs for home and visiting team
gamelogsdf[[10,11]].sum(axis=0)

# ^^ we add up all values in column 10 and 11 and we see that home teams outscored the visiting team 

# if we want to add up across columns (stay within row, we can use axis = 1)
# we can save this new data in total runs column 
gamelogsdf["totalRuns"] = gamelogsdf[[10,11]].sum(axis=1)
gamelogsdf["totalRuns"].head(20)
```

```python
%matplotlib inline 
#lets plot a distribution of total runs 
gamelogsdf["totalRuns"].hist(bins=30)
```

```python
import seaborn as sb 
sb.distplot(gamelogsdf[[10,11]].sum(axis=1),bins=30)
```

```python
#lets get rid of the funky kernel desnsity estimate 
sb.distplot(gamelogsdf[[10,11]].sum(axis=1),bins=30,kde=False)
```

```python
#looks like 9 runs happens the most frequently --> lets verify it with value counts 
gamelogsdf["totalRuns"].value_counts()
```

```python
gamelogsdf["totalRuns"].sum(axis=0)
gamelogsdf["totalRuns"].average(axis=0)
```

```python
# find the avg number of values in the first base umpires names?


import re 
```

```python
#we'd need to do this for all umpiress, we can write a function oursleves, or we can use a lambda function

# a lambda function is an anonymous function -- it is not named and used one time, typically for a small task.
```

```python
vowelCount = lambda x: len(re.findall("[aeiou]",x.lower()))
umpireName = "Jack Driscoll"
print(vowelCount(umpireName))
```

```python
# now lets use a lambda function on column 81
gamelogsdf[81].apply(lambda x: len(re.findall("[aeiou]",x.lower())))


# now we can get the average of vowels in the first base umpires name 
np.mean(gamelogsdf[81].apply(lambda x: len(re.findall("[aeiou]",x.lower()))))
```

```python
# we can even get a distribution the vowels
gamelogsdf[81].apply(lambda x: len(re.findall("[aeiou]",x.lower()))).hist(bins=5)
```
