```python
import numpy as np
```

```python
# boolean masking
# 3x3 array
threeByThree = np.array([[3,4,2],[6,8,1],[3,1,5]])
print(threeByThree)
```

```python
# which cells in the array are even?
# how do we figure out something is even?
# number % 2 is 0
print(threeByThree % 2)
# this gets us part of the way there
```

```python
print(threeByThree % 2 == 0)
# this comparison is called a "mask" 
# it is "hiding" the things we don't care about
# this this case, items in cells that are True we do care about, items in cells
# that are False we don't
```

```python
# we're going to use our mask with the bracket notation that numpy uses
# first, assign the mask to a variable
evenMask = threeByThree % 2 == 0
```

```python
print(evenMask)
```

```python
print(threeByThree)
print()
print(threeByThree[evenMask])
```

```python
oddMask = threeByThree % 2 == 1
print(oddMask)
```

```python
oddMask
```

```python
# create an array of only odd numbers
oddNumArray = threeByThree[oddMask]
```

```python
print(oddNumArray)
```

```python
type(oddNumArray)
```

```python
oddNumArray.shape # get rows and columns
```

```python
# you can use ANY expression in the mask
evenNumArray = threeByThree[evenMask]
```

```python
evenNumArray > oddNumArray # this doesn't work because the arrays are different shapes
```
