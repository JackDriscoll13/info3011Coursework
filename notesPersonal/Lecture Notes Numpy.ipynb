```python
#reason for numby
listOfNumbers =[10,9,8,7,6,5]
listOfNumbers
#to add one to each number we 
for i in listOfNumbers:
    listOfNumbers[10-i]+=1   # <----- rather complex way of doing it, is there a better way? 
print(listOfNumbers)
```

```python
#Numpy for arrays!
import numpy as np 
numpyArray = np.array(listOfNumbers)  #<--- puts list of numbers in specifified numpy array 
print(numpyArray)
numpyArray = numpyArray+1
print(numpyArray)

#can convert back to list (numpyObject.tolist())

listedArray = numpyArray.tolist()
print("\n{0}".format(listedArray))

```

```python
#Numpy methods:
#create new array: .array
myArray = np.array([1,2,3,4,5,6])

#arange: Works like range ---> np.arange(min,max,stepsize)
rangedArray = np.arange(100)
stepRangedArray = np.arange(30,51,2)
print(stepRangedArray)

#SHAPES AND TYPES 
myBadList = [3,4,"howdy"]
badArray = np.array(myBadList)  #<---- everything becomes a string! Yikes !

#.dtype is an np object property # gives u datatype of *data* in the array
print(badArray.dtype)  
print(stepRangedArray.dtype)
stepRangedArray*2   #becase we know type of array(int) here, we can multiple by 2 for specific arrray 

#Shape gives you the row, column dimensions of the array
a2DArray = np.array([[1,2],[3,4],[5,6]])
print(a2DArray)
print(a2DArray.shape)

#Get items from array:
print(a2DArray[2][0])
#NUMPY HAS SPECIAL ACCESSOR SYNTAX
print(a2DArray[1,0]) #uses commas instead of seperate brackets 
```

```python
#NUMPY Arrays contd.. slicing: 
print(a2DArray)
print()

#last row
print(a2DArray[2:])
#second row and below?
print(a2DArray[1:])
#First column
print(a2DArray[0:,0])
#second column
print(a2DArray[0:,1])
print(a2DArray[:,0])# this says give me every row but only the first column
```
