```python
# crosstabs
# we can use the crosstab function to calculate the cross tabulation of two variables
# column 13 specifies if a game was played at night or during the day
# column 3 specifies the day of the week
dayGametimeCrosstab = pd.crosstab(gamelogsDF[3],gamelogsDF[13])
dayGametimeCrosstab
```

```python
# can we order the days of the week?  we sure can!
orderedDays = ["Sun","Mon","Tue","Wed","Thu","Fri","Sat"]
dayGametimeCrosstab.loc[orderedDays]
```

```python
# we can also use multiple columns in our crosstabs
# if you didn't know, there are two leagues in baseball: NL, AL
# who plays the most night game?
pd.crosstab(gamelogsDF[3],gamelogsDF[8])
```

```python
dayGametimeLeagueCrosstab = pd.crosstab(gamelogsDF[3],[gamelogsDF[8],gamelogsDF[13]])
dayGametimeLeagueCrosstab
```

```python
dayGametimeLeagueCrosstab.loc[orderedDays]
```

```python
# we can do other things with crosstabs using different calculations. right now we're just 
# counting things up but we can also calculate averages, mins, maxs, medians, et cetera
dayAvgGameDurationByLeagueCrosstab = pd.crosstab(gamelogsDF[3],[gamelogsDF[8],
                                    gamelogsDF[13]],values=gamelogsDF[19],aggfunc=np.mean)
dayAvgGameDurationByLeagueCrosstab.loc[orderedDays]
```

```python
# export crosstab data as a csv file
dayAvgGameDurationByLeagueCrosstab.loc[orderedDays].to_csv("gameduration.csv")
```

```python
# reshaping data
# let's look at our weekday/league crosstab again
weekdayLeagueCrosstab = pd.crosstab(gamelogsDF[3],gamelogsDF[8])
weekdayLeagueCrosstab
```

```python
# reshape this dataframe
stackedWeekdayLeague = weekdayLeagueCrosstab.stack()
stackedWeekdayLeague
# the stack method turns columns into indexes
```

```python
type(stackedWeekdayLeague)
```

```python
# we want this data in a dataframes
# we can use unstack to turn indexes into columns
unstackedWeekdayLeague = weekdayLeagueCrosstab.unstack()
unstackedWeekdayLeague
# notice how the days of the week are in a column now
```

```python
resetUnstackedWeekdayLeague = unstackedWeekdayLeague.reset_index()
resetUnstackedWeekdayLeague
```

```python
resetUnstackedWeekdayLeague.columns = ["League","Day of Week","Total Games"]
resetUnstackedWeekdayLeague
```

```python
# why go through all of this trouble? this is an example of "tidy" data and is useful for
# plotting
orderedDays
```

```python
# introducing seaborn!
import seaborn as sb
```

```python
sb.catplot(x="Day of Week",y="Total Games",hue="League",
           data=resetUnstackedWeekdayLeague,kind="bar",order=orderedDays)
# catplot: is a plot used for categorical data (days of the week are categories)
# x: column name used for the x-axis
# y: column name used for the y-axis
# hue: splits the data up between/across multiple categories, each category gets its own color
# kind: specifies what type of chart (options: point, bar, count, box, violin, strip)
# order: order of the categorical data
```
