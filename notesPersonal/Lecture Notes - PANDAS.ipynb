```python
#datetime 
#datetime is a particular object type, where output format of datetime is 
# YYYY-mm-dd HH:MM:SS
# in our example we dont see HH:MM:SS ---> because they are 0 

#col 1 is the date of the game 
#pandas has functions for handling dates 
# we'll look at to_datetime   (CAPITILIZE THE Y -- for input)
gamelogsdf["datetime"]=pd.to_datetime(gamelogsdf[1],format="%Y%m%d")

#first parameter: source data
#second parameter: input format ----> if you specify an input format that doesnt match the actual input format. Pandas will throw an error
# if you dont specify format, pandas will guess, but doesnt always do a great job
#to_datetime returns new data -- it does not modify existing data 
```

```python
#strfttime can change your date output format 
gamelogsdf["datetime"].dt.strftime("%m/%d/%Y")
```

```python
#crosstabs
# we can use crosstab function to calculate the cross tabulation of two variables 
# column 13 specifies if the game was played at night 
# column 3 specifies the day of the week 
gamelogsdf.iloc[:,2:13].head(10)
```

```python
#crosstabs
#dayGametimeCrosstab = gamelogsdf.


    
```

```python
# we can also use multiple columns in our crosstab
# two leauges in baseball: NL, AL
# who plays the most night games 
pd.crosstab(gamelogsdf[3],gamelogsdf[8])
```

```python
dayGametimeLeaugeCrosstab = pd.crosstab(gamelogsdf[3],[gamelogsdf[8],gamelogsdf[13]])
dayGametimeLeaugeCrosstab
```

```python
dayGametimeLeaugeCrosstab.loc[["Sun","Mon","Tue","Wed","Thu","Fri","Sat"]]
```

```python
#we can do other things with crosstabs using different calculations
# right now we are just counting things up - we can do averages, mins, maxs, medians, etc..
dayAvgGameDurationByLeagueCrossTab = pd.crosstab(gamelogsdf[3],[gamelogsdf[8],gamelogsdf[13]],values=gamelogsdf[19],aggfunc=np.mean)
dayAvgGameDurationByLeagueCrossTab.loc[["Sun","Mon","Tue","Wed","Thu","Fri","Sat"]]
```

```python
#exporting 
dayAvgGameDurationByLeagueCrossTab.loc[["Sun","Mon","Tue","Wed","Thu","Fri","Sat"]].to_csv("game")
```
