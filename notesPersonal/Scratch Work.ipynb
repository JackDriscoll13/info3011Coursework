```python
import numpy as np
import pandas as pd
battingdf = pd.read_csv("batting.csv")
battingdf
```

```python
# lets get the rows 1000-1010, with columns playerID, yearID, AB, HR, RBI
batting1 = battingdf.loc[list(range(1000,1011)),["playerID","yearID","AB","HR","RBI"]]
# now rows 2000-2010 with same columns
batting2 = battingdf.loc[list(range(2000,2011)),["playerID","yearID","AB","HR","RBI"]]
# now rows 3000-3010 with same colulms
batting3 = battingdf.loc[list(range(3000,3011)),["playerID","yearID","AB","HR","RBI"]]
type(batting3)
```

```python
# now lets join the dataframes together
all3frames = [batting1,batting2,batting3]
result = pd.concat(all3frames)
result
```
